```python
#	Отримати список питань з файлу survey_results_schema (поле "qname")
# Щоб виконати це завдання, нам потрібно порівняти дві множини даних: назви всіх колонок, які є в основній таблиці з відповідями (df), та список ідентифікаторів питань з таблиці опису (df1).
# Метод інтерсекції (перетину) залишить лише ті назви, які є в обох списках одночасно.
# 1. Створюємо множину назв стовпців з основної таблиці (де відповіді)
columns_set = set(df.columns)
```

```python
# 2. Створюємо множину значень з поля "qname" таблиці схеми
# Отримуємо масив унікальних значень qname з таблиці схеми
# .unique() гарантує, що назви питань не будуть повторюватися
schema_questions_array = df1['qname'].unique()
# Перетворюємо масив у множину для подальшої інтерсекції
schema_questions_set = set(schema_questions_array)
```

```python
# 3. Знаходимо перетин (ті питання, що є і там, і там)
common_questions = columns_set.intersection(schema_questions_set)
```

```python
# 4. Виводимо результат (кількість та самі питання)
print(f"Знайдено {len(common_questions)} спільних питань.")
print(list(common_questions))
```

```python
#1. Створюємо підмножину даних
#Спочатку виділимо лише ті колонки, які є і в схемі, і в опитуванні:
df_common = df[list(common_questions)]
```

```python
#Перевірити наявність відповідей на ці питання у кожного респондента
# Тепер використаємо метод .notna() - True, якщо відповідь є, і False, якщо там порожньо
# Створюємо таблицю, де замість відповідей буде True (є відповідь) або False (пусто)
answers_check = df_common.notna()
```

```python
# Подивимось на результат для перших респондентів
answers_check.head()
```

```python
# Підраховуємо кількість респондентів без пропущених значень
# Створюємо таблицю, де видалено всі рядки, в яких є хоча б один NaN
clean_df = df_common.dropna()
```

```python
# Рахуємо кількість рядків у цій чистій таблиці
count_clean = clean_df.shape[0]
```

```python
# Завдання 3. Статистичний аналіз досвіду респондентів
```

```python
# 1. Середнє значення (mean) - сума всіх років поділена на кількість людей
mean_val = round(df['WorkExp'].mean())
```

```python
# 2. Медіана (median) - значення, що стоїть рівно посередині відсортованого списку
median_val = round(df['WorkExp'].median())
```

```python
# 3. Мода (mode) - значення, яке зустрічається найчастіше
# mode() повертає Series (список), тому беремо перший елемент [0]
mode_val = round(df['WorkExp'].mode()[0])
```

```python
# Виводимо результати
print(f"Міри центральної тенденції для WorkExp:")
print(f"Середнє значення: {mean_val} років")
print(f"Медіана: {median_val} років")
print(f"Мода: {mode_val} років")
```

```python
# Завдання 7. Географічний аналіз компенсації Python-розробників
```

```python
# Відфільтровуємо тих, хто працював з Python
df_python = df[df['LanguageHaveWorkedWith'].str.contains('Python', na=False)]
```

```python
# Щоб статистика була точною, нам потрібно видалити рядки, де респондент не вказав свою річну зарплату (стовпець ConvertedCompYearly).
# Видаляємо порожні значення (NaN) саме у стовпці із зарплатою
df_python = df_python.dropna(subset=['ConvertedCompYearly'])
```

```python
# 3. Групуємо за країною та рахуємо середню (mean) і медіанну (median) зарплату
py_progers_by_country = df_python.groupby('Country')['ConvertedCompYearly'].agg(['mean', 'median'])
```

```python
# 4. Сортуємо за медіаною, щоб побачити найприбутковіші країни зверху
py_progers_by_country = py_progers_by_country.sort_values(by='median', ascending=False)
```

```python
# Округлюємо до 1 знака після коми
py_progers_by_country = py_progers_by_country.round(1)
```

```python
# *Завдання 9. Аналіз популярності Python по віковим категоріям 
```

```python
# 1. Створюємо прапор (True/False), чи працює людина з Python
df['IsPython'] = df['LanguageHaveWorkedWith'].str.contains('Python', na=False)
```

```python
# 2. Групуємо за віком та рахуємо:
# - загальну кількість респондентів у групі (count)
# - кількість тих, у кого IsPython == True (sum)
age_stats = df.groupby('Age')['IsPython'].agg(['count', 'sum'])
```

```python
# 3. Розраховуємо відсоток
age_stats['Percentage'] = (age_stats['sum'] / age_stats['count']) * 100
```

```python
# 4. Округляємо для охайності та сортуємо за відсотком (від популярніших до менш)
age_stats = age_stats.round(2).sort_values(by='Percentage', ascending=False)
```
